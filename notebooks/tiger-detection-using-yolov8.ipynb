```python
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import os
```

```python
class TigerDetector:
    
    def __init__(self, yaml_path: str, yolo_ckpt: str, epochs: int):
        self.yaml_path = yaml_path
        self.yolo_ckpt = yolo_ckpt
        self.model = YOLO(self.yolo_ckpt)
        self.epochs = epochs
    
    def __train_detector(self):
        self.model.train(data=self.yaml_path, epochs=self.epochs)
        metrics = self.model.val()
    
    def __evaluate(self):
        self.model.val(data=self.yaml_path, split="test")
        
    def train_eval(self):
        self.__train_detector()
        self.__evaluate()
```

```python
td = TigerDetector(yaml_path = "/kaggle/input/tiger-detection-dataset/final_data/data.yaml",
                   yolo_ckpt = "yolov8m.pt",
                   epochs = 50)
```

```python
td.train_eval()
```
